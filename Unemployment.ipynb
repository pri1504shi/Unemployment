```python
data.columns=['Region','Date','Freq','Est Unemp Rate','Est Emp','Est Lab Part Rate','Area']
data
```

```python
data['Region'].nunique()
```

```python
grouped_data = data.groupby('Region').size().reset_index(name='Count')
grouped_data
```

THIS IS THE ESTIMATED UEMPLOYMENT RATE ACCORDING TO THE DATE IN ALL THE STATES.

```python
sns.barplot(x=data['Est Unemp Rate'], y=data['Date'], data=data, palette= 'dark')
plt.show()
```

THIS IS A BAR PLOT BETWEEN ALL THE STATES AND SUM OF ESTIMATED EMPLOYED PEOPLE ON THE BASIS OF THE AREA (RURAL OR URBAN).

```python
df= data.groupby(['Region','Area'])['Est Emp'].sum().reset_index()
sns.barplot(x=df['Region'], y=df['Est Emp'], hue=df['Area'], palette='winter')
plt.xticks(rotation=90)
plt.show()
```

This is the pie chart between Mean Estimated Labour Participation Rate (%) and Area of the place.

```python
from ipywidgets.widgets.widget_string import Label
df=data.groupby(['Area'])['Est Lab Part Rate'].mean().reset_index()
df.plot(kind='pie',y='Est Lab Part Rate', labels= df['Area'], autopct='%1.1f%%',legend=False, colors='cyan')                    #0= rural , 1= urban
plt.show()
```

This is the maximum Unemployment Rate in each states in India.

```python
sns.set_theme()

df= data.groupby(['Region', 'Area'])['Est Unemp Rate'].median().reset_index()
colors= {'Rural':'Black', 'Urban':'Red'}

sns.scatterplot(x=df['Region'], y=df['Est Unemp Rate'], hue=df['Area'], palette= colors)
plt.grid(True, linewidth= 1.5)
plt.xticks(rotation=90)
plt.show()
```
